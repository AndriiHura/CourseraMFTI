```python
import math
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return math.sin(x/5.)*math.exp(x/10.) + 5*math.exp(-x/2.)
```

```python
x_coords = np.arange(1, 15, 0.1)
y_coords = np.array(list(map(lambda x: func(x), x_coords)))
```

```python
plt.plot(x_coords,y_coords)
```

## Polynomial of degree 1

```python
x_coords_1 = [[1,1], [1, 15]]
funcs_answers_1 = [func(1), func(15)]
w_coeffs_1 = linalg.solve(x_coords_1, funcs_answers_1)
print(w_coeffs_1)
```

```python
func1 = lambda x: w_coeffs_1[0] + x*w_coeffs_1[1]
```

```python
y_coords1 = np.array(list(map(lambda x: func1(x), x_coords)))
```

```python
plt.plot(x_coords,y_coords)
plt.plot(x_coords,y_coords1)
```

## Polynomial of degree 2

```python
x_coords_2 = [[1,1,1], [1, 8, 8**2], [1, 15, 15**2]]
funcs_answers_2 = [func(1), func(8), func(15)]
w_coeffs_2 = linalg.solve(x_coords_2, funcs_answers_2)
print(w_coeffs_2)
```

```python
func2 = lambda x: w_coeffs_2[0] + w_coeffs_2[1]*x + w_coeffs_2[2]*(x**2)
```

```python
y_coords2 = np.array(list(map(lambda x: func2(x), x_coords)))
```

```python
plt.plot(x_coords,y_coords)
plt.plot(x_coords,y_coords1)
plt.plot(x_coords,y_coords2)
```

## Polynomial of  degree 3

```python
x_coords_3 = [[1,1,1,1], [1,4,4**2,4**3], [1,10,10**2,10**3], [1,15,15**2,15**3]]
funcs_answers_3 = [func(1), func(4), func(10), func(15)]
w_coeffs_3 = linalg.solve(x_coords_3, funcs_answers_3)
print(w_coeffs_3)
```

```python
func3 = lambda x: w_coeffs_3[0] + w_coeffs_3[1]*x + w_coeffs_3[2]*(x**2) + w_coeffs_3[3]*(x**3)
```

```python
y_coords3 = np.array(list(map(lambda x: func3(x), x_coords)))
```

```python
plt.plot(x_coords,y_coords)
plt.plot(x_coords,y_coords1)
plt.plot(x_coords,y_coords2)
plt.plot(x_coords,y_coords3)
```

```python
result = [round(x, 2) for x in w_coeffs_3]
```
